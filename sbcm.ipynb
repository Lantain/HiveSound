```python
#https://www.kaggle.com/datasets/annajyang/beehive-sounds?select=all_data_updated.csv
import pandas as pd
import os
from src.dataset import segments_from_audio_file
```

```python
df = pd.read_csv('./dataset/archive (1)/all_data_updated.csv', header=1)
df
```

```python
df_subset = df[['queen status', 'file name']]
df_subset
```

```python
df_subset = df_subset.drop(df_subset[df_subset['queen status'].isin([1, 2])].index)
df_subset
```

```python
ddf = df_subset.sort_values('file name')
unique_file_names = df_subset['file name'].unique()
unique_file_names.size
```

```python
SRC_DIR = './dataset/archive (1)/sound_files/sound_files'
OUT_DIR = "./dataset/sbcm_full"
files = os.listdir(SRC_DIR)
for index, row in df_subset.iterrows():
    # Access row values using row['column_name']
    queen_status = row['queen status']
    file_name = row['file name'].replace('.raw', '')
    
    related_files = list()
    for f in files:
        if f.startswith(file_name):
            related_files.append(f)

    print(f"Queen Status: {queen_status}, File Name: {file_name}, Files: {related_files}")
    for rf in related_files:
        if queen_status == 0:
            os.makedirs(f"{OUT_DIR}/noqueen", exist_ok=True)
            segments = segments_from_audio_file(f"{SRC_DIR}/{rf}", 4000, 200, None)
            for i, seg in enumerate(segments):
                seg.set_sample_width(2)
                seg.export(f"{OUT_DIR}/noqueen/{rf}_{i}.wav", format="wav", bitrate='16k', parameters=["-sample_fmt", "s16"])
        if queen_status == 3:
            os.makedirs(f"{OUT_DIR}/queen", exist_ok=True)
            segments = segments_from_audio_file(f"{SRC_DIR}/{rf}", 4000, 200, None)
            for i, seg in enumerate(segments):
                seg.set_sample_width(2)
                seg.export(f"{OUT_DIR}/queen/{rf}_{i}.wav", format="wav", bitrate='16k', parameters=["-sample_fmt", "s16"])
```
