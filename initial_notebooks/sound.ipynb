```python
import os # библиотека для работы с файлами
import librosa
import matplotlib.pyplot as plt
import librosa.display as ld
import numpy as np
import IPython

OSBH_FILE_QUEEN = "dataset/osbh/queen/Hive1_12_06_2018_QueenBee_H1_audio___15_00_00_10.wav"
OSBH_FILE_QUEENLESS = "./dataset/osbh/no_queen/Hive1_31_05_2018_NO_QueenBee_H1_audio___15_10_00_0.wav"
SBCM_FILE_QUEEN = "./dataset/sbcm/queen/2022-06-13--18-08-05_1__segment0.wav_3.wav"
SBCM_FILE_QUEENLESS = "dataset/sbcm/noqueen/2022-06-05--17-41-01_2__segment0.wav_5.wav"
```

```python
def visualize(path):
    # Waveform
    signal, sr = librosa.load(path, sr = 22050) # загружаем файл
    signal
    plt.figure(figsize=(12,4))
    ld.waveshow(signal, sr=sr, color='b')

    #Show audio
    display(IPython.display.Audio(signal, rate = sr))

    # Spectrum
    n_fft = 2048
    ft = np.abs(librosa.stft(signal[:n_fft], hop_length = n_fft+1))
    plt.plot(ft)
    plt.title('Spectrum')
    plt.xlabel('Frequency Bin')
    plt.ylabel('Amplitude')

    # Spectrogram
    X = librosa.stft(signal)
    s = librosa.amplitude_to_db(abs(X))
    ld.specshow(s, sr=sr, x_axis = 'time', y_axis='linear')
    plt.colorbar()
```

```python
visualize(OSBH_FILE_QUEEN)
```

```python
visualize(OSBH_FILE_QUEENLESS)
```

```python
mfccs = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc = 40, hop_length=512)
mfccs
```

```python
melspectrum = librosa.feature.melspectrogram(y=signal, sr = sr,
                                        	hop_length =512, n_mels = 40)
librosa.display.specshow(melspectrum, sr=sr, x_axis='time', y_axis='mel')
```

```python
cent = librosa.feature.spectral_centroid(y=signal, sr=sr)
plt.figure(figsize=(15,5))
plt.semilogy(cent.T, label='Spectral centroid')
plt.ylabel('Hz')
plt.legend()
```

```python
rolloff = librosa.feature.spectral_rolloff(y=signal, sr=sr)
plt.figure(figsize=(15,5))
plt.semilogy(rolloff.T, label='Roll-off frequency')
plt.ylabel('Hz')
plt.legend()
```

```python
zrate=librosa.feature.zero_crossing_rate(signal)
plt.figure(figsize=(14,5))
plt.semilogy(zrate.T, label='Fraction')
plt.ylabel('Fraction per Frame')
plt.legend()
```
