```python
import tensorflow as tf
from src.dataset import stft_dataset, OSBH_PATH
from src.audio import to_spectrogram_dataset

train_stft_ds, val_stft_ds, label_names, input_shape = stft_dataset(OSBH_PATH, 0.2, 12)
```

# Model (OSBH + STFT + CONV) training

```python
from src.model import get_conv_model, callbacks
from src.plots import plot_history
import tensorboard


LEARNING_RATE = 0.0001

model = get_conv_model(input_shape, len(label_names))
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

history = model.fit(
    train_stft_ds,
    validation_data=train_stft_ds,
    epochs=30,
    callbacks=[callbacks]
)
plot_history(history)
```

```python
def history_to_summary(prefix: str, history):
    print(f"{prefix}:    Loss: {history.history['loss'][-1]:.2f}, Acc: {history.history['accuracy'][-1]:.2f}, Val. Loss: {history.history['val_loss'][-1]:.2f}, Val. Acc: {history.history['val_accuracy'][-1]:.2f}" )

history_to_summary("OSBH + STFT + Conv model    ", history)
model.save('models/osbh_stft_conv.keras')
```
